```python
import pandas as pd
import streamlit as st
import plotly.express as px
```

```python
df = pd.read_csv('C:/Users/Dr_Na/Git_Repositories/Practicum_Project/vehicles_us.csv')
display(df.head(10))
```

```python
avg_price = int(df['price'].mean())
median_price = int(df['price'].median())
print (f'The average price of cars is : ${avg_price}')
print (f'The median price of cars is: ${median_price}')
```

```python
avg_days_listed = int(df['days_listed'].mean())
print (f'The average number of days a car is listed: {avg_days_listed}')
```

```python
median_year = int(df['model_year'].median())
print(f'The most common model of car is from: {median_year}')
```

```python
excellent_list = []
fair_list = []
good_list = []
like_new_list =[]
new_list = []
salvage_list = []

for index in range(len(df)):
    if df['condition'][index] == 'excellent':
        excellent_list.append(df['condition'][index])
    if df['condition'][index] == 'fair':
        fair_list.append(df['condition'][index])
    if df['condition'][index] == 'good':
        good_list.append(df['condition'][index])
    if df['condition'][index] == 'like new':
        like_new_list.append(df['condition'][index])
    if df['condition'][index] == 'new':
        new_list.append(df['condition'][index])    
    if df['condition'][index] == 'salvage':
        salvage_list.append(df['condition'][index])    
        
print(f'The total number of cars in excellent condition is: {len(excellent_list)}')
print(f'fair condition: {len(fair_list)}')
print(f'good condition: {len(good_list)}')
print(f'like_new condition: {len(like_new_list)}')
print(f'new condition: {len(new_list)}')
print(f'salvage condition: {len(salvage_list)}')
```
